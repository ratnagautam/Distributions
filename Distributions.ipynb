4.What is the concept of standardization (z-score), and why is it important in data analysis? Explain the
formula and how standardization transforms a dataset.

    -Standardization (Z-score) in Data Analysis

    Standardization, also known as a z-score, is a statistical method used to convert data into a common scale. It shows how far a particular value is from the mean of the dataset in terms of standard deviations. This concept is an important part of Statistics and is widely used with the Normal Distribution.

    Explanation
    Standardization transforms the original data into a new scale where the mean becomes 0 and the standard deviation becomes 1. After transformation:

    A positive z-score means the value is above the mean
    A negative z-score means the value is below the mean

    For example, a z-score of +2 means the value is 2 standard deviations above the mean.

    Importance of Standardization
    *It helps in comparing values from different datasets
    *It makes it easier to identify outliers
    *It simplifies data analysis and interpretation
    *It is widely used in statistical methods and  research
    
    Conclusion
    Standardization is a useful technique that converts data into a standard form, making it easier to compare, analyze, and interpret values across different datasets.

5.What is Kurtosis and their type?

    -Kurtosis is a measure in Statistics that describes the shape of a distribution, particularly the peakedness and heaviness of the tails compared to a normal distribution.

    It helps us understand whether the data has extreme values (outliers) or is more evenly spread.

    Types of Kurtosis
    a)Mesokurtic
    This is the normal distribution type.
    It has a moderate peak and normal tails.
    Example: Normal Distribution
    Kurtosis value ≈ 3

    b)Leptokurtic
    Has a high peak and heavy (fat) tails
    Indicates more outliers in the data
    Kurtosis value > 3

    c)Platykurtic
    Has a flat peak and light tails
    Indicates fewer outliers
    Kurtosis value < 3

6.Explain why the uniform distribution is a good model for the outcome of rolling a fair die.

    -A Uniform Distribution is a distribution in which all outcomes are equally likely.

    When a fair die is rolled, there are six possible outcomes: 1, 2, 3, 4, 5, and 6. Since the die is fair, each outcome has the same probability:

    P(1)=P(2)=P(3)=P(4)=P(5)=P(6)=1/6

    Why Uniform Distribution is a Good Model
    a)Equal Probability of Outcomes
    Each number on a fair die has an equal chance of occurring, which is the main property of a uniform distribution.

    b)No Bias or Preference
    A fair die does not favor any particular number, so the outcomes are evenly distributed.

    c)Discrete and Finite Outcomes
    The die has a fixed number of outcomes (6), and each is equally likely, making it a perfect example of a discrete uniform distribution.

    P(X≥8)

    Given:
    Number of trials n=15
    Probability of success p=0.5
    Required: P(X≥8)

    In Microsoft Excel, use:

    =1 - BINOM.DIST(7, 15, 0.5, TRUE)

    Explanation:

    BINOM.DIST(7, 15, 0.5, TRUE) gives P(X≤7)
    So, subtracting from 1 gives P(X≥8)
    P(X≥8)≈0.5

8.How does log transformation help in stabilizing variance and making data more normally distributed?

    -Log Transformation in Data Analysis

    Log transformation is a technique used in Statistics to improve the distribution of data by applying a logarithm (usually log base 10 or natural log) to each value.

    How Log Transformation Stabilizes Variance:
    In many datasets, especially those with large values, the variability increases as the values increase (called heteroscedasticity).

    *Log transformation reduces the spread of large values more than small values
    *This makes the variance more constant across the dataset
    *As a result, the data becomes easier to analyze using statistical methods

    How It Makes Data More Normally Distributed:
    *Many real-world datasets are right-skewed (long tail on the right)
    *Log transformation compresses high values and reduces skewness
    *This helps the data resemble a Normal Distribution (bell-shaped curve)
